```python
import pandas as pd
import numpy as np
import re 
import nltk
import string

# from 
from collections import Counter
from nltk.corpus import stopwords
```

```python
nltk.corpus.stopwords
```

```python
working_df = pd.read_csv('masterFile.csv')
working_df= working_df.dropna()
```

```python
all_categories = working_df['category'].values,
unique_cat = np.unique(all_categories)
```

```python
def text_parser(block, cnt):
    #Better text parsing code here
    
    shortword = re.compile(r'\W*\b\w{1,3}\b')
    block = shortword.sub('', block)
    
    lst_of_tokens = block.split(' ')
    
    lst_of_tokens = [w.lower() for w in lst_of_tokens]
    
    table = str.maketrans('', '', string.punctuation)
    lst_of_tokens = [w.translate(table) for w in lst_of_tokens]
    
    lst_of_tokens = [token for token in lst_of_tokens if token.isalpha()]
    
    stop_words = set(stopwords.words('english'))
    lst_of_tokens = [w for w in lst_of_tokens if not w in stop_words]
    lst_of_tokens.sort()
    
    extra_sw = ['officer', 'police', 'department']
    lst_of_tokens = [w for w in lst_of_tokens if not w in extra_sw]
    lst_of_tokens.sort()
    
    officer_names = []
    lst_of_tokens = [w for w in lst_of_tokens if not w in officer_names]
    lst_of_tokens.sort()
    
    for i in lst_of_tokens:
        cnt[i] +=1
    return cnt
```

```python
selected_cols_df = working_df[['category','text_content']]
category_dict = dict()
category_counter_dict = dict()
for category in unique_cat:
    category_counter_dict[category] = Counter()
    
for i in range(len(selected_cols_df.values)): 
    #iterating through all the rows in the selected cols
    #Have text content, with its associated category
    category = selected_cols_df.values[i][0]
    content = selected_cols_df.values[i][1]
    category_counter_dict[category] = text_parser(content, category_counter_dict[category]) 
    #Do a frequency count of the content.
    #Should return the top used words.

for category in unique_cat:
    if category in category_dict:
        for val in category_counter_dict[category].most_common(20):
            category_dict[category].append(val)
    else:
        category_dict[category] = []
        for val in category_counter_dict[category].most_common(20):
            category_dict[category].append(val)
```
